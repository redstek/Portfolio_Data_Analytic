```python
##############################################   Задание 1.3   #################################################################

##### Добавьте на визуализацию из прошлого пункта вертикальную линию красного цвета в точке 6 
##### Добавьте также пунктирную вертикальную линию черного цвета в точке 2
##### Сделайте кривую y(x) зеленого цвета
##### Добавьте легенду для максимума и минимума в правый нижний угол


plt.figure(figsize=(7, 7))

plt.plot(x, y, color='g', label='Кривая y(x)')
plt.axvline(x=6, color='r', label='Линия в точке 6')
plt.axvline(x=2, linestyle='--', color='k', label='Линия в точке 2')

max_idx = y.index(max(y))
min_idx = y.index(min(y))

plt.scatter(x[max_idx], y[max_idx], color='red', label='Максимум')
plt.scatter(x[min_idx], y[min_idx], color='blue', label='Минимум')


plt.title("Третий график")
plt.xlabel("Ряд Х")
plt.ylabel("Ряд Y")

plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

```python
##############################################   Задание 1.4   #################################################################

##### Добавьте на график еще одную кривую z = [6,9,13,13,16,12,20,18,17] и сделайте ее фиолетовго цвета
##### Теперь должно быть 4 вертикальные линии - max(Y), min(Y), max(Z), min(Z). 
##### Подпишите их на легенде и выберите цвета для них.Максимумы должны быть пунктирные, а минимумы цельные


plt.figure(figsize=(7, 7))

plt.plot(x, y, 'g', x, z, 'm')

max_y_idx = y.index(max(y))
min_y_idx = y.index(min(y))
max_z_idx = z.index(max(z))
min_z_idx = z.index(min(z))

plt.scatter(x[max_y_idx], y[max_y_idx], color='r')
plt.scatter(x[min_y_idx], y[min_y_idx], color='b')
plt.scatter(x[max_z_idx], z[max_z_idx], color='r')
plt.scatter(x[min_z_idx], z[min_z_idx], color='b')

plt.axvline(x[max_y_idx], color='red', linestyle='--', label=' линия Максимума Y')
plt.axvline(x[min_y_idx], color='blue', label=' линия Минимума Y')
plt.axvline(x[max_z_idx], color='purple', linestyle='--', label=' линия Максимума Z')
plt.axvline(x[min_z_idx], color='orange', label=' линия Минимума Z')

plt.title("Четвёртый график")
plt.xlabel("Ряд Х")
plt.ylabel("Ряд Y")

plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

```python
##############################################   Задание 1.6   #################################################################

##### Добавьте на правую ось столбчатую диаграмму, которая будет показывать разницу между Y и Z
##### Сделайте эту диаграмму полупрозрачной с помощью аргумента alpha
##### Добавьте подпись для диаграммы на отдельный элемент легенды.
##### Расположите основную легенду справа-снизу, а легенду вспомогательной оси справа-сверху

########  Hint: используйте twinx() для создания дополнительной оси


fig, ax = plt.subplots(figsize=(7, 7))

ax.plot(x, y, 'g', x, z, 'm')

max_y_idx = y.index(max(y))
min_y_idx = y.index(min(y))
max_z_idx = z.index(max(z))
min_z_idx = z.index(min(z))

ax.scatter(x[max_y_idx], y[max_y_idx], color='r')
ax.scatter(x[min_y_idx], y[min_y_idx], color='b')
ax.scatter(x[max_z_idx], z[max_z_idx], color='r')
ax.scatter(x[min_z_idx], z[min_z_idx], color='b')

ax.axvline(x[max_y_idx], color='red', linestyle='--', label=' линия Максимума Y')
ax.axvline(x[min_y_idx], color='blue', label=' линия Минимума Y')
ax.axvline(x[max_z_idx], color='purple', linestyle='--', label=' линия Максимума Z')
ax.axvline(x[min_z_idx], color='orange', label=' линия Минимума Z')

ax2 = ax.twinx()

difference = [y_val - z_val for y_val, z_val in zip(y, z)]
ax2.bar(x, difference, alpha=0.1, label='Разница (Y - Z)')

ax.set_title("Шестой график")
ax.set_xlabel("Ряд Х")
ax.set_ylabel("Ряд Y")

ax.grid(True)

ax.legend(loc='lower right')
ax2.legend(loc='upper right')

plt.show()
```

```python
##############################################   Задание 2.1   #################################################################

##### Импортните csv-файл под названием Purchase_List
##### Приведите date_purchase к формату даты с помощью функции to_datetime
##### Создайте колонку датафрейма с месяцесем оплат (числом, от 1 до 12)
##### Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
##### Постройте визуализацию: для каждого месяца покупки по левой оси постройте сумму оплат, а по правой - количество оплат
##### Не забудьте оформление графика (заглавие, подпись осей, раздельная легенда, сетка)

purchase_list['date_purchase'] = pd.to_datetime(purchase_list['date_purchase'])

purchase_list['month'] = purchase_list['date_purchase'].dt.month

grouped_df = purchase_list.groupby('month').agg({'id_purchase': 'count', 'amt_pay': 'sum'}) \
    .rename(columns={'id_purchase': 'Количество оплат', 'amt_pay': 'Сумма оплат'})
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Месяц')
ax1.set_ylabel('Сумма оплат', color=color)
g_first = ax1.plot(grouped_df.index, grouped_df['Сумма оплат'], color='r', label='Сумма оплат')


ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Количество оплат', color=color)
g_second = ax2.plot(grouped_df.index, grouped_df['Количество оплат'], color='m', label='Количество оплат')

lines = g_first + g_second
labels = [line.get_label() for line in lines]

plt.title('Статистика оплат по месяцам')
plt.xlabel('Месяц')

plt.grid(True)
plt.legend(lines, labels, loc='upper right')

plt.show()
```

```python
##############################################   Задание 2.2   #################################################################

##### Постройте scatter plot для среднего чека для каждого месяца.
##### Для этого создайте новую колонку в сгруппированном датафрейме с отношением суммы платежей к их количеству
##### Сохраните форматирование из предыдущих пунктов
##### Задайте нижнюю границу оси Y как ноль. Верхнюю границу придумайте самостоятельно для лучшего отображения информации.

grouped_df['Средний чек'] = grouped_df['Сумма оплат'] / grouped_df['Количество оплат']
```

```python

fig, ax = plt.subplots()

ax.scatter(grouped_df.index, grouped_df['Средний чек'])

ax.set_title('Средний чек по месяцам')
ax.set_xlabel('Месяц')
ax.set_ylabel('Средний чек')

ax.grid(True)
ax.set_ylim(bottom=0, top=4000)

plt.show()
```
